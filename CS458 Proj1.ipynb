```python
import numpy as np
import math 
import matplotlib.pyplot as plt

# Function calculates distance between two points 
def dist(p1, p2, d): 
      
    x = []
    for i in range(d):
        x.append(p1[i] - p2[i])
        
    tot = 0
    for i in range(d):
        tot = tot + (x[i] * x[i])
        
    return tot
  
# Function to find the max dist between any 2 pnts 
def maxDist(p, d): 
  
    n = len(p) 
    maxm = 0
  
    # Iterate over all possible pairs 
    for i in range(n): 
        for j in range(i + 1, n): 
              
            # Update maxm 
            maxm = max(maxm, dist(p[i], p[j], d)) 
  
    # Return actual distance 
    return sqrt(maxm) 
      
# Function to find the min dist between any 2 pnts
def minDist(p, d):
    
    n = len(p)
    minm = 10
    
    # Iterate over al possible pairs
    for i in range(n):
        for j in range(i + 1, n):
            
            # Update minm
            minm = min(minm, dist(p[i], p[j], d))
            
    # Return actual distance        
    return sqrt(minm)

xaxis = []
yaxis = []
i = 2

# Generate 1000 points following a uniform distribution under 2-50 dimensions
while i <= 50:
    p = []
    j = 0
    while (j < 1000):
        point = np.random.uniform(0,1,i)
        p.append(point)
        j = j + 1
        
    # Find max distance between any 2 points    
    maxd = maxDist(p, i)
    
    # Find min distance between any 2 points
    mind = minDist(p, i)
    
    # Keep track of dimension number
    xaxis.append(i)
    
    # Keep track of difference between max and min distance for given dimension
    yaxis.append(math.log10((maxd - mind) / mind))
    
    i = i + 1

# Plot log10((max-min)/min) under 2-50 dimensions
plt.plot(xaxis, yaxis)
```

```python
from sklearn import datasets
iris = datasets.load_iris()
import matplotlib.pyplot as plt
from sklearn.preprocessing import KBinsDiscretizer
from pandas import DataFrame
import pandas as pd

# store feature matrix in "X"
X = iris.data[:, 2:]

# store response vector in "y"
Y = iris.target

# create petal length x petal width graph
plt.figure(2, figsize=(8, 8))
plt.xlabel('Petal length')
plt.ylabel('Petal width')

# list of lists to keep track of all points from each class
kmeans = [[0,0], [0,0], [0,0]]

# plot points 
for i in range(len(Y)):
    if Y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='red', edgecolor='k', s=30)
        kmeans[0][0] += X[i][0]
        kmeans[0][1] += X[i][1]
    elif Y[i] == 1:
        plt.scatter(X[i][0], X[i][1], color='green', edgecolor='k', s=30)
        kmeans[1][0] += X[i][0]
        kmeans[1][1] += X[i][1]
    else:
        plt.scatter(X[i][0], X[i][1], color='blue', edgecolor='k', s=30)
        kmeans[2][0] += X[i][0]
        kmeans[2][1] += X[i][1]
        
# calculate kmean for each class
for i in range(len(kmeans)):
    kmeans[i][0] = kmeans[i][0]/50
    kmeans[i][1] = kmeans[i][1]/50

# calculate where discretizing lines should go
box1 = [(kmeans[0][0]+kmeans[1][0])/2, (kmeans[0][1]+kmeans[1][1])/2]
box2 = [(kmeans[1][0]+kmeans[2][0])/2, (kmeans[1][1]+kmeans[2][1])/2]

# plot discretizing lines
plt.plot(1.5, 'k--')
plt.axhline(y=box1[1], xmin=0., xmax=(0.5), linestyle='--', c='black')
plt.axvline(x=box1[0], ymin=0., ymax=(box1[1]/(box2[1]/1.4)), linestyle='--', c='black')
plt.axhline(y=box2[1], xmin=box2[0], xmax=(0.435), linestyle='--', c='black')
plt.axhline(y=box1[1], xmin=0., xmax=(box2[0]/6.8), linestyle='--', c='black')
plt.axvline(x=box2[0], ymin=box1[1], ymax=(0.3), linestyle='--', c='black')
plt.axvline(x=box2[0], ymin=box1[1], ymax=(box2[1]), linestyle='--', c='black')

# list flower type and the number of times they were correctly or incorrectly separated in table
table_data = [[50,48,45],[0,2,5]]
columnNames = ['Setosa','Versicolor','Virginica']
rowNames = ['Correct', 'Incorrect']
datatable = DataFrame(table_data, rowNames, columnNames)
display(datatable)
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

iris = datasets.load_iris()

# store feature matrix in "X"
X = iris.data

# store response vector in "y"
y = iris.target

# create a sepal length x sepal width graph of the iris data
plt.figure(2, figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
        
# use PCA to reduce dimensions from 4 to 2
X_reduced = PCA(n_components=2).fit_transform(X)

# plot all samples after PCA
fig, ax = plt.subplots(1, 1, figsize=(6,6))
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
```
